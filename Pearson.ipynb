```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr
from numpy.linalg import eig


def pearson_function(item1, item2):
    corr, _ = pearsonr(item1, item2)
    return corr

def correlation_visualization(item1, item2, df):
    sns.scatterplot(x = item1, y = item2, data = df)  # to visualise correlation between variables
    corr = df.corr()                                  # to visualise correlation matrix
    corr = corr.style.background_gradient(cmap='Blues')
    return corr

def calculate_zscore(df):
    standardized_data = (df - df.mean())/(df.std())
    return standardized_data

def calculate_covariance_eigen(df):
    covariance_matrix = np.cov(df, bias=True)
    print("Covariance_Matrix")
    print(covariance_matrix)
    eigen_values, eigen_vectors = eig(covariance_matrix)
    idx = eigen_values.argsort()[::-1]
    eigen_values = eigen_values[idx]
    eigen_vectors = eigen_vectors[:,idx]
    print("Eigen Values")
    print(eigen_values)
    print('Eigen Vectors')
    print(eigen_vectors)
    
    
    

df = pd.read_csv('pearson/iris.csv')
item1 = df['petal_length']
item2 = df['petal_width']
print('Pearsons correlation: %.3f' % pearson_function(item1, item2))
df = df.drop(columns = ['species'])
standardized_df = calculate_zscore(df)
print("Standardized Data using Z-Score")
print(standardized_df)
calculate_covariance_eigen(standardized_df)
correlation_visualization(item1, item2, df)








    

```
