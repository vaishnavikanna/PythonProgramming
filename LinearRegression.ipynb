```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train_data = pd.read_csv('train.csv')
traindf = pd.DataFrame(train_data)
df = traindf.dropna(axis = 0, how='any')



trX = df['x'].values  #training samples
trY = df['y'].values
trX_mean = np.mean(trX)
trY_mean = np.mean(trY)

a = 0
b = 0
for i in range(len(trX)):
    a += (trX[i] - trX_mean)*(trY[i] - trY_mean)  
    b += (trX[i] - trX_mean)**2
slopeTrain = a / b            #slope of training data
interTrain = trY_mean - (slopeTrain*trX_mean)  #intercept of training data

print("Slope and Intercept of training data")
print (slopeTrain, interTrain)
testdata = pd.read_csv('test.csv')
testdf = pd.DataFrame(testdata)
testdf = testdf.dropna()
testX = testdf['x'].values
testY = testdf['y'].values
predicted_label = slopeTrain*testX + interTrain  #predicted values with test data

print("Predicted continuous labels for test data")
print(predicted_label)
meanSquaredError = np.square(np.subtract(testY,predicted_label)).mean() 
 
rootMeanSquaredError = np.sqrt(meanSquaredError)  #rmse error metric showing efficiency of prediction
print("Root Mean Squared Error")
print(rootMeanSquaredError)



```
