```python
import requests
import json
from bs4 import BeautifulSoup
import urllib
from urllib.parse import urlparse, urljoin
import pandas as pd


class parse_webscience:

    def parse_internal(self, page_url):
        internal_urls = []
        external_urls = []
        samepage_urls = []
        internal_count = 0
        external_count = 0
        samepage_count = 0
        page_count = 0
        urlrequest = requests.get(url=page_url)

        df = pd.DataFrame()
        domain_name = urlparse(page_url).netloc

        soup = BeautifulSoup(urlrequest.text, 'html.parser')
        # df = pd.DataFrame(columns=['INTcount', 'EXTcount', 'URLFragments', 'timestamp'])
        if soup.find('script', type='application/ld+json'):
            jsonstr = soup.find('script', type='application/ld+json').text
            jsonObj = json.loads(jsonstr)
            date_string = jsonObj['dateModified']
            print(date_string)
        else:
            date_string = None
            print(date_string)

        for link in soup.find_all("a"):
            href = link.get("href")
            href = urljoin(page_url, href, allow_fragments=True)
            parsed_href = urlparse(href, scheme='', allow_fragments=True)
            fragmenthref = ""
            if str(parsed_href.fragment):
                fragmenthref = str(parsed_href.scheme) + "://" + str(parsed_href.netloc) + str(parsed_href.path) + "#" + str(parsed_href.fragment)
            else:
                href = str(parsed_href.scheme) + "://" + str(parsed_href.netloc) + str(parsed_href.path)

            if domain_name in href:
                if fragmenthref:
                    if fragmenthref not in samepage_urls:
                        # print(f"same page link: {fragmenthref}")
                        samepage_urls.append(fragmenthref)
                        samepage_count+=1
                        continue
                else:
                    if href not in internal_urls:
                        print(f"Internal Link: {href}")
                        internal_urls.append(href)
                        internal_count+=1
                        continue

            if domain_name not in href:
                if href not in external_urls:
                    # print(f"External Link: {href}")
                    external_urls.append(href)
                    external_count+=1
                    continue
            if len(internal_urls) > 20:
                break
            page_count += 1

            #data = {"INTcount": len(internal_urls), "EXTcount": len(external_urls), "URLFragments": len(samepage_urls),
                                # "timestamp": date_string}

            #df = df.append(data)

        # df_res1 = pd.DataFrame(index=page_count)
        #df_res1 = df
        print(len(internal_urls))
        print(len(external_urls))
        print(len(samepage_urls))
        return internal_urls






p = parse_webscience()
url_to_parse = 'https://simple.wikipedia.org/wiki/Climate_change'
internal_urls = p.parse_internal(url_to_parse)
#print(df_result)

for each_link in internal_urls:
    print(f"parsed link: {each_link}")
    p.parse_internal(each_link)
#print(df_result)
```
